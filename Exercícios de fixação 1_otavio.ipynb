1) Ler uma data (dia, mês e ano) e informar quantos dias se passaram desde o início do ano informado.

Considere:

* A quantidade correta de dias em cada mês
* Considerar anos bissextos segundo as seguinte regras:
   * Um ano é bissexto se for divisível por 4
   * Exceto os anos divisíveis por 100
   * Porém, os anos divisíveis por 400 continuam sendo bissextos
* Validação da data antes de fazer o cálculo

```python
dia: 27
mes: 3
ano: 2026

dia = int(input("Dia:"))
mes = int(input("Mês:"))
ano = int(input("Ano:"))

#verifica se o ano é bissexto
bissexto = (ano % 4 == 0 and ano % 100 !=0) or (ano % 400 == 0)

#verifica a quantidade de dias de cada mês
dias_mes = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

#ajusta fevereiro se for bissexto
if bissexto:
    dias_mes[1] = 29

#validação da data. Perceba que a fração do "código dia > dias_mes[mes - 1]" puxa a informação da posição do mês na lista, se mês abril, está na posição [3] da lista, por isso [4 - 1]. 
#o trecho do código "sum(dias_mes[:mes - 1]) + dia" baiscamente diz para somar a quantidade de dias de cada mês especificado na lista "dias_mes" até (:) a posição especificada (4-1 = 3 -> mês de abril).
if mes < 1 or mes > 12:
    print ("Mes informado invalido")
elif dia < 1 or dia > dias_mes[mes - 1]:
    print("Dia informado invalido")
else: 
    dias_passados = sum(dias_mes[:mes - 1]) + dia
    print(f"Já se passaram {dias_passados} dias desde o início do ano")
```
